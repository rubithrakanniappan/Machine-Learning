```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import lightgbm as lgb
 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#Here we load the dataset from the 50_Startups.csv file
dataset = pd.read_csv("insurance.csv")
#Check the dataset data in table format
dataset
```

```python
#pandas.get_dummies() is used for data manipulation. It converts categorical data into dummy or indicator variables.
#drop_first=True - Remove first level to get n-1 dummies out of n categorical level
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
# it returns the column labels of the dataset
dataset.columns
```

```python
#Split the inputs
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
#Split the output
dependent=dataset[["charges"]]
dependent
```

```python
#Train and Test data split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#import AdaBoostRegressor from sklearn.ensemble
from sklearn.ensemble import AdaBoostRegressor
#import make_regression from sklearn.datasets
from sklearn.datasets import make_regression
#Assigned AdaboostRegressor to regressor variable
regressor= AdaBoostRegressor(n_estimators=100, loss='linear', random_state=0)
#Here is our model creation happening
regressor=regressor.fit(X_train,Y_train)
```

```python
#check the evaluate metrics
#this is used for cross check the values
#call predict function using regressor and pass the X_test as input 
y_pred=regressor.predict(X_test)
```

```python
#find the R Square value so import r2_score from sklearn.metrics 
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
#R_score value between 0 to 1. If it is nearly to 1 then this is a good model
#our r_score is 0.93 so it is perform/learn well 
r_score
```

```python
#This is to save the model for later use
import pickle 
#Create a filename to save model. .sav is the extention of pickle to save file
filename="Finalized_model_AdaBoost.sav"
```

```python
#insert the model into Finalized_model_linear.sav file
#if we save this, then we can able to use this later very easily 
#'wb' - indicates that the file is opened for writing in binary mode
pickle.dump(regressor, open(filename,'wb'))
```

```python
#Load the saved model to check whether it is working as expected before go for phase 2(Deployment)
#'rb' - indicates that the file is opened for read in binary mode
loaded_model=pickle.load(open("Finalized_model_AdaBoost.sav",'rb'))
#Pass the year of experience in the predict function
#save the model and predict the value through loaded model
#We have to pass 5 inputs in the predict function
result=loaded_model.predict([[1234,345,4565,1,0]])
#View the profit details in the result variable
result
```
