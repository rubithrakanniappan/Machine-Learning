```python
#Pandas is an open source library in Python. This is made mainly for working relational or labeled data. 
#It is fast and efficient for manipulating and analyzing data. 
#import the all functions under the pandas in this file memory.
#pd is the alias name of pandas. whereever we call pandas we can use pd 
#import all function from pandas so we have memory consumption 
import pandas as pd 
```

```python
#Here we load the dataset from the 50_Startups.csv file
dataset = pd.read_csv("50_Startups.csv")
```

```python
#Check the dataset data in table format
dataset
```

```python
#pandas.get_dummies() is used for data manipulation. It converts categorical data into dummy or indicator variables.
#drop_first=True - Remove first level to get n-1 dummies out of n categorical level
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
# it returns the column labels of the dataset
dataset.columns
```

```python
#Split the inputs
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
independent
```

```python
#Split the output
dependent=dataset[["Profit"]]
dependent
```

```python
#Train and Test data split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#import DecisionTreeRegressor from sklearn.tree
#upgrade the scitit-learn using this line - pip install -U scikit-learn in jupyter notebook itself
from sklearn.tree import DecisionTreeRegressor
#Assigned DecisionTreeRegressor to regressor variable
#Hyper tuning parameters are defined as the parameters that are explicitly defined by the user to control the learning process
regressor=DecisionTreeRegressor(criterion='absolute_error', splitter='best', max_features='auto')
#Here only our model creation happening
regressor=regressor.fit(X_train,Y_train)
```

```python
#Matplotlib is a plotting library for creating static, animated, and interactive visualizations in Python
import matplotlib.pyplot as plt
#import tree from sklearn
from sklearn import tree
#it plot a decision tree
tree.plot_tree(regressor)
#we can view the decision tree
plt.show()
```

```python
#check the evaluate metrics
#this is used for cross check the values
#call predict function using regressor and pass the X_test as input 
y_pred=regressor.predict(X_test)
```

```python
#find the R Square value so import r2_score from sklearn.metrics 
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
#R_score value between 0 to 1. If it is nearly to 1 then this is a good model
#our r_score is 0.93 so it is perform/learn well 
r_score
```

```python
#This is to save the model for later use
import pickle 
#Create a filename to save model. .sav is the extention of pickle to save file
filename="Finalized_model_DecisionTree.sav"
```

```python
#insert the model into Finalized_model_linear.sav file
#if we save this, then we can able to use this later very easily 
#'wb' - indicates that the file is opened for writing in binary mode
pickle.dump(regressor, open(filename,'wb'))
```

```python
#Load the saved model to check whether it is working as expected before go for phase 2(Deployment)
#'rb' - indicates that the file is opened for read in binary mode
loaded_model=pickle.load(open("Finalized_model_DecisionTree.sav",'rb'))
#Pass the year of experience in the predict function
#save the model and predict the value through loaded model
#We have to pass 5 inputs in the predict function
result=loaded_model.predict([[67565,45679,76589,0,1]])
#View the profit details in the result variable
result
```
