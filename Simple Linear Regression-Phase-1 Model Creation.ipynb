```python
#Pandas is an open source library in Python. This is made mainly for working relational or labeled data. 
#It is fast and efficient for manipulating and analyzing data. 
#import the all functions under the pandas in this file memory.
#pd is the alias name of pandas. whereever we call pandas we can use pd 
#import all function from pandas so we have memory consumption 
import pandas as pd 
```

```python
#Here we load the dataset from the Salary_Data.csv file
#For that we call pd(Pandas-class name) + read_csv(function name) 
#read_csv needs the variable parameter so we have to pass the csv file name in double quotation
#Finally dataset variable get all data from Salary_Data.csv file
#Requried file should be place respective folder and also use exact filename otherwise we will get 404 error that mean file not found in the path.
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#Dataset variable get all data from Salary_Data.csv file
#we can see the data in the table format using dataset variable 
dataset
```

```python
#Input Split 
#Get the YearsExperience column data in independent variable from dataset 
independent = dataset[["YearsExperience"]]
#check the get data using independent 
independent
```

```python
#Output Split 
#Get the Salary column data in dependent variable from dataset
dependent = dataset[["Salary"]]
#Check the get data using dependent 
dependent
```

```python
#Train and Test data split 
#import the train_test_split class using filename (syntax: from filename import classname) 
#.(Dot) operator using to call the respective class or function (Example: from Foldername.Filename import classname)
#what is the different between direct import and this import. This type of import will helps to save memory.
from sklearn.model_selection import train_test_split
#It is the parameterised function so we pass the parameter without any spelling mistake to avoid unwanted error when run 
#this fuction returning the 4 values so we declare the 4 variable like below
#we can't give space when variable declaration so we used _ inbetween X and train and same for rest of variable
#Variable name is the userdefined but position is pre-defined. 
#1st - input train position, 2nd - input test position, 3rd - output train position and 4th - output test position 
X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#This is the model creation/Learning Process
#import LinearRegression class using sklearn.linear_model folder and filename
from sklearn.linear_model import LinearRegression
#assign LinearRegression class to regressor variable
regressor=LinearRegression()
#using regressor call fit function and pass the train input and output value. 
#Here only we can find the weight and bias value 
regressor.fit(X_train,Y_train)
```

```python
#Check the weight value
weight=regressor.coef_
weight
```

```python
#Check the bias value
bias=regressor.intercept_
bias
```

```python
#check the evaluate metrics
#this is used for cross check the values
#call predict function using regressor and pass the X_test as input 
y_pred=regressor.predict(X_test)
```

```python
#find the R Square value so import r2_score from sklearn.metrics 
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
#R_score value between 0 to 1. If it is nearly to 1 then this is a good model
#our r_score is 0.97 so it is perform/learn well 
r_score
```

```python
#This is to save the model for later use
import pickle 
#Create a filename to save model. .sav is the extention of pickle to save file
filename="Finalized_model_linear.sav"
```

```python
#insert the model into Finalized_model_linear.sav file
#if we save this, then we can able to use this later very easily 
#'wb' - indicates that the file is opened for writing in binary mode
pickle.dump(regressor, open(filename,'wb'))
```

```python
#Load the saved model to check whether it is working as expected before go for phase 2(Deployment)
#'rb' - indicates that the file is opened for read in binary mode
loaded_model=pickle.load(open("Finalized_model_linear.sav",'rb'))
#Pass the year of experience in the predict function
#save the model and predict the value through loaded model
result=loaded_model.predict([[15]])
#View the salary details in the result variable
result
```
